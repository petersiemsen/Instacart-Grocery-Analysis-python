```python
# Function to see the count of all the missing values in our dataframe df_products.
# We can see that product_name has 16 counts of missing data, or NaN. 

df_products.isnull().sum()
```

```python
# Function to actually view those 16 rows from the previous function. 
# It creates a subset called df_nan and adds any rows where the product_name is null.

df_nan = df_products[df_products['product_name'].isnull() == True]
```

```python
df_nan
```

```python
df_products.shape
```

```python
# Creating a new dataframe for our "clean" product data. We are using the .isnull() == False
# to make sure to grab all the rows that have complete data in them for product_name. The rows 
# that have NaN in product_name will not be included. 

df_products_clean = df_products[df_products['product_name'].isnull() == False]
```

```python
df_products_clean.shape
```

```python
# Function to look for any FULL DUPLICATES in our clean dataframe.

df_dups = df_products_clean[df_products_clean.duplicated()]
```

```python
df_dups

# After looking at the subset, we can see that there are 5 total rows
# that have complete duplicate data. 
```

```python
# Dropping the duplcates with the .drop_duplciates() function. We've created a new
# subsetted dataframe to hold the new rows.

df_products_clean_no_dups = df_products_clean.drop_duplicates()
```

```python
df_products_clean_no_dups

# We see that the new dataframe has 5 less rows (49,672 total)
```

```python
# Exporting our new, cleaned product data into a .csv file

df_products_clean_no_dups.to_csv(os.path.join(path, '02. Data', 'Prepared Data', 'products_checked.csv'))
```

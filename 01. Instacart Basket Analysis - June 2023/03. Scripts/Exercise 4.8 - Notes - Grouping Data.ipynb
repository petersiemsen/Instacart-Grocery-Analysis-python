```python
# Importing the orders_products_ pkl file

df_orders_products_merged = pd.read_pickle(os.path.join(path, '02. Data', 'Prepared Data', 'orders_products_busiest_days.pkl'))
```

```python
# Creating subset of just the first 1,000,000 rows

df = df_orders_products_merged[:1000000]
```

```python
# Grouping the dataframe by product name

df.groupby('product_name')
```

```python
# Using the aggregation function to find the mean order_number, grouped by department_id

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Running multiple stats on the column 'order_number'

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Create a new column called 'max_order'. We are then going to look at each user_id and find the max
# order_number from each user's order history.

df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_orders_products_merged[['user_id', 'max_order']]
```

```python
# Using the loc() function to find values in our dataframe, in the 'max_order' column with certain criteria.
# From there, we are creating a new column called 'loyalty_flag' and assigning each a string value.

df_orders_products_merged.loc[df_orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['max_order'] <= 40) & (df_orders_products_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_orders_products_merged.loc[df_orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking out new columns to see the results.

df_orders_products_merged[['user_id', 'max_order', 'loyalty_flag']]
```

```python
# value_counts for the new loyalty_flag

df_orders_products_merged['loyalty_flag'].value_counts(dropna = False)
```

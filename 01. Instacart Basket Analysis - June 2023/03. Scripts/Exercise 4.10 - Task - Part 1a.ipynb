### 3. Comparing the customers behavior based on geographic areas.

#### We will be creating a regional segmentation of the data. A new column called "reguion" will be added to the dataset and will be based on the customers State.

```python
# Creating a list called 'region' based on location criteria

region = []

for value in df_merged['state']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
# Convert the list to a pandas Series

region_series = pd.Series(region)

# Use value_counts() to get the count of each unique value in the Series

region_series.value_counts()
```

```python
# Adding the list 'region' to our df_merged dataframe

df_merged['region'] = region
```

```python
# Checking results to see that the column was added 

df_merged.head()
```

### Determine if there is a difference in spending habits between the different U.S. regions. 

#### We are going to use a crosstab to compare the two variables. 

```python
# Creating crosstab 

crosstab_region = pd.crosstab(df_merged['region'], df_merged['spend_flag'], dropna = False)
```

```python
# Adding in a % column of 'High spender' out of the total spending population

crosstab_region['High Spender %'] = crosstab_region.apply(lambda row: row['High spender'] / (row['High spender'] + row['Low spender']) * 100, axis=1)
```

```python
# Adding in a % column of 'Low spender' as a percentage of the total spending population

crosstab_region['Low Spender %'] = crosstab_region.apply(lambda row: row['Low spender'] / (row['High spender'] + row['Low spender']) * 100, axis=1)
```

```python
crosstab_region
```

#### We can see that the 'South' states have the highest proportion of 'Low Spenders' compared to the other states. The Northeast has the lowest number of Low Spenders. Proportionally though, the Midwest has the highest % of High Spenders compared to its total spender population. The Northeast has the lowest. 

### 4. Exclude any customers who have less than 5 orders. The CFO isn't interested in these cusomters in the analysis. 

#### In order to do this, we will need to create a flag for each customer based on the max_order column. If it is < 5, we will exclude them from the altered dataset. 

```python
# Creating a 'customer_activity' column and 'High-activity' for >= 5 max orders

df_merged.loc[df_merged['max_order'] >= 5, 'customer_activity'] = 'High-activity'
```

```python
# Creating the 'Low-activity' exclusion flag for < 5 max orders

df_merged.loc[df_merged['max_order'] < 5, 'customer_activity'] = 'Low-activity'
```

```python
# Checking the results to see that the columns were added 

df_merged.value_counts('customer_activity')
```

```python
# Creating a subset of the data for only the 'High-activity' customers so we can export them

df_high_activity = df_merged[df_merged['customer_activity'] == 'High-activity']
```

```python
# Checking to see that it was created correctly

df_high_activity.shape
```

```python
# Exporting the df_high_activity subset so that we can analyze it further in a new notebook

df_high_activity.to_pickle(os.path.join(path, '02. Data', 'Prepared Data', 'high_activity_customers.pkl'))
```

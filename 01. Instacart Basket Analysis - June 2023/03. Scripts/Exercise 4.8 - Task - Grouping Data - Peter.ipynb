#### Question 4: Create a loyalty flag for existing customers based on their number of orders. 

```python
# First, we need to create a column called 'max_order' to see the highest number of orders each customer has made.

df_orders_products_merged2['max_order'] = df_orders_products_merged2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking our results 

df_orders_products_merged2
```

```python
# Using the loc() function to find values in our dataframe, in the 'max_order' column with certain criteria.
# From there, we are creating a new column called 'loyalty_flag' and assigning each a string value.

df_orders_products_merged2.loc[df_orders_products_merged2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_orders_products_merged2.loc[(df_orders_products_merged2['max_order'] <= 40) & (df_orders_products_merged2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_orders_products_merged2.loc[df_orders_products_merged2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking out new columns to see the results.

df_orders_products_merged2[['user_id', 'max_order', 'loyalty_flag']]
```

#### Question 5: The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. 

```python
# Running the agg() function with multiple statistics on the 'prices' column. We want to group by
# loyalty_flag so that we can compare the differences between the 3 groups.

df_orders_products_merged2.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

#### Question 6: Create a spending flag for each user based on the average price across all their orders. Mean price < 10, flag them as "Low spender". Mean price >= 10, flag them as "High spender". 

```python
# First, we need to create a new column called 'mean_spend' that is grouped by the user_id. This way, we can compare the
# average amount they spend across all their orders. 

df_orders_products_merged2['mean_spend'] = df_orders_products_merged2.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking our results

df_orders_products_merged2.head()
```

```python
# The next step is to use the loc() function to run through each row and assign the named flags based on the new 
# mean_spend column we just created. 

df_orders_products_merged2.loc[df_orders_products_merged2['mean_spend'] < 10, 'spend_flag'] = 'Low spender'
```

```python
df_orders_products_merged2.loc[df_orders_products_merged2['mean_spend'] >= 10, 'spend_flag'] = 'High spender'
```

```python
# Checking our results with a value_counts() to see how it was distributed. 

df_orders_products_merged2['spend_flag'].value_counts(dropna = False)
```

#### Question 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# First, we need to create a new column called 'median_days_since' that is grouped by the user_id. This way, 
# we can compare the median number of days since their last order.

df_orders_products_merged2['median_days_since'] = df_orders_products_merged2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking our results

df_orders_products_merged2.head()
```

```python
# The next step is to use the loc() function to run through each row and assign the named flags based on the new 
# median_days_since column we just created. 

df_orders_products_merged2.loc[df_orders_products_merged2['median_days_since'] > 20, 'order_behavior_flag'] = 'Non-frequent customer'
```

```python
df_orders_products_merged2.loc[(df_orders_products_merged2['median_days_since'] > 10) & (df_orders_products_merged2['median_days_since'] <= 20), 'order_behavior_flag'] = 'Regular customer'
```

```python
df_orders_products_merged2.loc[df_orders_products_merged2['median_days_since'] <= 10, 'order_behavior_flag'] = 'Frequent customer'
```

```python
# Checking our results with a value_counts() to see how it was distributed. 

df_orders_products_merged2['order_behavior_flag'].value_counts(dropna = False)
```

```python
# Check for NaN values in the 'median_days_since' column. I wanted to investigate these rows a little more based on our
# value_counts() from above. 

nan_columns = df_orders_products_merged2[df_orders_products_merged2['median_days_since'].isnull()]

# Display the columns with NaN values

nan_columns
```
